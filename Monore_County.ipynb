```python
def from_date():
    """ TO GET 30 DAYS BACK DATE """
    try:
        now = datetime.now()
        date_str = datetime(now.year, now.month - 2, now.day - 1).strftime('%m/%d/%Y')
        return date_str
    except Exception as e:
        print("Error at from_date() = ",e)


def to_date():
    """ TO GET CURRENT DATE """
    try:
        now = datetime.now()
        date_str = datetime(now.year, now.month, now.day - 1).strftime('%m/%d/%Y')
        return date_str
    except Exception as e:
        print("Error at to_date() = ",e)
```
